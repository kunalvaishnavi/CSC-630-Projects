# Continuation of Maps Project by Kunal Vaishnavi

I decided for my open project, I wanted to upgrade some features in my maps project, specifically the following:

- Let the user choose the airline
- Let the user choose the route
- Let the user choose the airport
- Program the airline, route, and time series through onclicks of buttons
- Program the airport through an input box and a submit button

Though these aren't many features to add on, they take a lot of time and can be considered as their own project. 

To do the first thing, I first created the buttons for what I wanted and designed their looks as well. Here was the first button I created:

Then, I created other buttons that were dropdown menus and let the user choose what kind of visualization he/she wants to see:

Next, I had to bind those dropdown menus to functions that would update the data and then display the correct graph. To do this, I created one huge function, which sorted out what kind of dropdown was selected, loaded the corresponding data, and fed that data to another function, which handled displaying the data.

The above two pictures of code took many hours to figure out because the data was originally plotting only January data. In the end, it was a silly mistake of mine. I was calculating startPos incorrectly.

Now that was done, I began creating the customizability of origin and destination airports. However, I realized that making a dropdown wouldn't be feasible, since there's so many airports. Instead, I decided to create a box where the user could type the airport's IATA code (the three letter abbreviation) and get the visualization accordingly.

This also took several hours because I was trying to figure out how to use these input boxes without submitting a form and there wasn't any easy way. This [link](https://developer.mozilla.org/en-US/docs/Web/HTML/Element/input/search) helped me understand what I was doing wrong. Afterwards, I cleaned up my code and added some finishing touches such as carets on the dropdowns using this [link](http://lugolabs.com/articles/6-create-a-thin-caret-with-html-and-css).

One thing that I wanted to do but did not have time to figure out how to combine all of the filters in the format I currently have. The only design that I saw that would enable me to combine all of the filters was with a checkbox list where you preselect everything and then graph, something that I didn't want to do.
